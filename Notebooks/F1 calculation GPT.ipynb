```python
import pandas as pd
from sklearn.metrics import f1_score, precision_score, accuracy_score, recall_score
from statistics import mean

languages = ['english', 'chinese', 'turkish', 'spanish', 'yoruba', 'polish', 'ukrainian']

table = {'language': [], 'f1': [], 'precision': [], 'recall': [], 'f1bin':[], 'precbin':[], 'recbin':[], 'accuracy': []}

for i in languages:
    f1 = []
    prec = []
    rec = []
    acc = []
    f1bin = []
    precbin = []
    recbin = []
    for j in range(0,5):
        a = pd.read_csv(i+'/0shot_prompt2_'+i+'_'+str(j)+'.csv')
        
        for k in a.index:
            if a.loc[k]['response'] not in [0,1,'0','1']:
                a.drop(k, inplace = True)
        
        f1.append(f1_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        prec.append(precision_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        rec.append(recall_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        f1bin.append(f1_score(a['true_value'].astype(int),a['response'].astype(int)))
        precbin.append(precision_score(a['true_value'].astype(int),a['response'].astype(int)))
        recbin.append(recall_score(a['true_value'].astype(int),a['response'].astype(int)))
        acc.append(accuracy_score(a['true_value'].astype(int),a['response'].astype(int)))
    table['language'].append(i)
    table['f1'].append(mean(f1))
    table['precision'].append(mean(prec))
    table['recall'].append(mean(rec))
    table['accuracy'].append(mean(acc))
    table['f1bin'].append(mean(f1bin))
    table['precbin'].append(mean(precbin))
    table['recbin'].append(mean(recbin))
```
