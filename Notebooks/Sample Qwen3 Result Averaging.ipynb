```python
import pandas as pd
from sklearn.metrics import f1_score, precision_score, accuracy_score, recall_score
from statistics import mean
import re

languages = ['english', 'chinese', 'turkish', 'spanish', 'ukrainian']
languages = ['chinese']

table = {'language': [], 'f1': [], 'precision': [], 'recall': [], 'accuracy': []}
rep = """<think>

</think>

"""

for i in languages:
    f1 = []
    prec = []
    rec = []
    acc = []
    for j in range(0,5):
        a = pd.read_csv('0shot_prompt2_qwen_'+i+'_'+str(j)+'.csv')
             
        a['response'] = a['response'].apply(lambda x: re.sub(rep,'',x))
        a.to_csv(i+'_qwen_fewshot_'+str(j)+'.csv')
        #print(a['response'].head())
        for k in a.index:
            if a.loc[k]['response'] not in [0,1,'0','1']:
                a.drop(k, inplace = True)
        
        f1.append(f1_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        prec.append(precision_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        rec.append(recall_score(a['true_value'].astype(int),a['response'].astype(int), average='macro'))
        acc.append(accuracy_score(a['true_value'].astype(int),a['response'].astype(int)))
    table['language'].append(i)
    table['f1'].append(mean(f1))
    table['precision'].append(mean(prec))
    table['recall'].append(mean(rec))
    table['accuracy'].append(mean(acc))
```
